### What is "Boston" dataset?

The Boston Housing Dataset is a classic dataset used in the field of machine learning and statistics, particularly for regression analysis. It contains information about the housing market in various Boston suburbs, primarily focusing on housing prices and relevant features that could potentially affect these prices. The dataset is widely used for educational and testing purposes in predictive modeling. Here is a brief overview of its attributes:

1. **CRIM**: Per capita crime rate by town.
2. **ZN**: Proportion of residential land zoned for lots over 25,000 sq.ft.
3. **INDUS**: Proportion of non-retail business acres per town.
4. **CHAS**: Charles River dummy variable (1 if tract bounds river; 0 otherwise).
5. **NOX**: Nitric oxides concentration (parts per 10 million).
6. **RM**: Average number of rooms per dwelling.
7. **AGE**: Proportion of owner-occupied units built prior to 1940.
8. **DIS**: Weighted distances to five Boston employment centres.
9. **RAD**: Index of accessibility to radial highways.
10. **TAX**: Full-value property-tax rate per $10,000.
11. **PTRATIO**: Pupil-teacher ratio by town.
12. **B**: 1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town.

13. **LSTAT**: Percentage of lower status of the population.
14. **MEDV**: Median value of owner-occupied homes in $1000s.

The dataset is primarily used for regression tasks, where "MEDV" (median value of owner-occupied homes) often serves as the target variable. The other attributes provide various insights into factors that might influence housing prices, such as crime rates, environmental conditions, and the social-economic status of the population.

It's important to note that some attributes, like "B" and "LSTAT", have been subjects of ethical discussions due to their potential for reinforcing biases.

# **3. Regression: Apply "Data Preprocessing"**
---

### **3.1 Model performance using "Normalization"**

```python
setup(data = bostonDataSet, target = 'medv',
      normalize = True, normalize_method = 'zscore', silent=True)
cm = compare_models()

#normalize_method = {zscore, minmax, maxabs, robust}
```

---
### **3.2 Model performance using "Feature Selection"**
---

```python
setup(data = bostonDataSet, target = 'medv',
      feature_selection = True, feature_selection_method = 'classic',
      feature_selection_threshold = 0.2,  silent=True)
cm = compare_models()

#feature_selection_method = {classic, boruta}
```

---
### **3.3 Model performance using "Outlier Removal"**
---

```python
setup(data = bostonDataSet, target = 'medv',
      remove_outliers = True, outliers_threshold = 0.05, silent=True)
cm = compare_models()
```

---
### **3.4 Model performance using "Transformation"**
---

```python
setup(data = bostonDataSet, target = 'medv',
      transformation = True, transformation_method = 'yeo-johnson', silent=True)
cm = compare_models()
```

---
### **3.5 Model performance using "PCA"**
---

```python
setup(data = bostonDataSet, target = 'medv',
      pca = True, pca_method = 'linear', silent=True)
cm = compare_models()
```

```python
setup(data = bostonDataSet, target = 'medv',
      remove_outliers = True, outliers_threshold = 0.05,
      normalize = True, normalize_method = 'zscore', silent=True)
cm = compare_models()
```
